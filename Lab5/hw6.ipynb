## Weather Scraping
Create a class called Weather to scrape weather.com data. Initialize Weather to take in a zipcode. Implement the following attributes and methods (many of which are completed already):
#### Attributes
- self.zip returns the zipcode
- self.soup returns the html page parsed by beautiful soup

#### Methods
- get_url() # returns the weather.com url based on the zip code entered
- get_html() # returns the page's html source code
- get_soup() # returns the pages html source code parsed by beautiful soup
- feels_like() # returns the "feels like" temperature without any characters (use the strip() method)
- today() # returns today's temperature without any characters (use the strip() method)
- high() # returns today's high without any characters (use the strip() method)
- low() # returns today's low without any characters (use the strip() method)
- wind() # returns today's windspeed without any characters (use the strip() method)
- humidity() # returns today's humidity without any characters (use the strip() method)
- dewpoint() # returns today's dewpoint without any characters (use the strip() method)
- pressure() # returns today's pressure without any characters (use the strip() method)
- visibility() # returns today's visibility without any characters (use the strip() method)
- csv_destination_file_name() # returns the file name to write to, which will be the self.zip + ".csv"
- header_data() # identifies the header data and takes an file name argument
- write_header() # writes the header data to the first row of the csv file
- to_csv() # outputs the data to a csv file

For all methods, just store the numbers (not %, mph, NE etc.). There are many ways to scrape the information, but the BeautifulSoup methods: .find(), .findall(), .text() and .strip(), as well as using the list-like structure of each element scraped can accomplish it all.

```python
import urllib.request as url
from bs4 import BeautifulSoup

class Weather:
    def __init__(self, zip):
        super(Weather, self).__init__() 
        self.zip = str(zip)
        self.soup = self.get_soup()

    def get_url(self):
        return "https://weather.com/weather/today/l/" + self.zip + ":4:US"
    
    def get_html(self): 
        local_filename = url.Request(self.get_url(), headers={'User-Agent': 'Mozilla/5.0'})
        return url.urlopen(local_filename).read()
    
    def get_soup(self):
        html = self.get_html()
        return BeautifulSoup(html, "html.parser")
    
    def feels_like(self):
        # Hint: look for span tag and class deg-feels in self.soup # Your code here
        feelsLike = self.soup.find_all('span',{'class':'deg-feels'}) 
        l = feelsLike[0].get_text().strip()
        return l.replace("°","")
    
    def today(self):
        # First step to scrape the high and low from today. We search for s # Your code here
        today = self.soup.find_all('span',{'class':'deg-hilo-nowcard'}) 
        return today
    
    def high(self):
        # Hint: use the method self.today() above and index the result to g # Your code here
        today = self.today()
        high = today[0].get_text().strip()
        return high.replace("°","")
        
    def low(self):
        # Hint: use the method self.today() above and index the result to g # Your code here
        today = self.today()
        low = today[1].get_text().strip()
        return low.replace("°","")
        
    def wind(self):
        wind = self.soup.find_all('td',{'class':''}) 
        with_units_wind = wind[0].get_text().strip() 
        wind_speed = ''
        for i in with_units_wind:
            if not i.isalpha() and not i.isspace():
                wind_speed = wind_speed + i 
        return wind_speed
    
    def humidity(self):
        # Hint search self.soup attribute for td tag and class "" and index # Your code here
        humidity = self.soup.find_all('td',{'class':''}) 
        with_units_humid = humidity[1].get_text().strip()
        Humidity = ''
        for i in with_units_humid:
            if not i.isalpha() and not i.isspace():
                Humidity = Humidity + i
        return Humidity.replace("%","")
    
    def dewpoint(self):
        # Hint search self.soup attribute for td tag and class "" and index # Your code here
        dewpoint = self.soup.find_all('td',{'class':''}) 
        with_units_dewpoint = dewpoint[2].get_text().strip()
        dewp = ''
        for i in with_units_dewpoint:
            if not i.isalpha() and not i.isspace():
                dewp = dewp + i
        return dewp.replace("°","")
    
    def pressure(self):
        # Hint search self.soup attribute for td tag and class "" and index # Your code here
        pressure = self.soup.find_all('td',{'class':''}) 
        with_units_pressure = pressure[3].get_text().strip()
        Pressure = ''
        for i in with_units_pressure:
            if not i.isalpha() and not i.isspace():
                Pressure = Pressure + i
        return str(round(float(Pressure)))
    
    def visibility(self):
        # Hint search self.soup attribute for td tag and class "" and index # Your code here
        vis = self.soup.find_all('td',{'class':''})
        with_units_vis = vis[4].get_text().strip()
        vis = ''
        for i in with_units_vis:
            if not i.isalpha() and not i.isspace():
                vis = vis + i
        return str(round(float(vis)))
    
    def csv_destination_file_name(self): 
        return self.zip + ".csv"
    
    def header_data(self, f): 
        return ",".join([
        'zip',
        'feel',
        'high',
        'low',
        'wind (mph)',
        'humidity (%)',
        'dewpoint (degrees)',
        'pressure (inches)',
        'visibility miles'
])
    def write_header(self, f): 
        f.write(self.header_data(f)) 
        f.write("\n") # \n is a newline character
        
    def write_data(self, f): 
        f.write(",".join([
            self.zip,
            self.feels_like(),
            self.high(),
            self.low(),
            self.wind(),
            self.humidity(),
            self.dewpoint(),
            self.pressure(),
            self.visibility()
        ]))
        f.write("\n")
        
    def to_csv(self):
        f = open(self.csv_destination_file_name(), "w") 
        self.write_header(f)
        self.write_data(f)
```

```python
# Instantiate two instances of the class Weather SLO(93405) and Harlem(10026)
SLO = Weather(93405)
Harlem = Weather(10026)
```

```python
# Test the method to_csv
SLO.to_csv()
Harlem.to_csv()
```
